```python
#Data Cleaning
#ðŸ”¸ Missing & Null Values

data.isnull().sum()
```

```python
# Fill missing postal codes with mode (most common postal code)
data['Postal Code'] = data['Postal Code'].fillna(data['Postal Code'].mode()[0])
data['Postal Code']
```

```python
data.isnull().sum()
```

```python
#ðŸ”¸ Duplicates
data.duplicated().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualizing duplicate rows count
duplicates = data.duplicated().sum()

plt.figure(figsize=(4, 4))
sns.barplot(x=['Unique Rows', 'Duplicate Rows'], y=[len(data)-duplicates, duplicates])
plt.title('Duplicate Data Overview')
plt.show()
```

```python
#ðŸ”¸ Data Types & Formats
# Convert Order Date and Ship Date columns to datetime format
data['Order Date'] = pd.to_datetime(data['Order Date'], dayfirst=True)
data['Order Date'].head()
```

```python
data['Ship Date'] = pd.to_datetime(data['Ship Date'], dayfirst=True)
data['Ship Date'] .head()
```

```python
data.dtypes
```

```python
# Calculate duration (in days)
data['Shipping Delay'] = (data['Ship Date'] - data['Order Date']).dt.days

# Display few rows to verify
data[['Order Date', 'Ship Date', 'Shipping Delay']].head()
```

```python
# Convert categorical columns
categorical_cols = ['Ship Mode', 'Segment', 'Region', 'Category', 'Sub-Category']
data[categorical_cols] = data[categorical_cols].astype('category')
data[categorical_cols].dtypes
```

```python
# Convert Postal Code to integer (handle missing safely)
data['Postal Code'] = data['Postal Code'].astype('Int64')
data['Postal Code'] .dtypes
```

```python
data.dtypes
```

```python
#ðŸ”¸ String / Text Cleaning
text_cols = ['Customer Name', 'Product Name', 'Category']

for col in text_cols:
    data[col] = data[col].str.strip()          # start/end spaces remove
    data[col] = data[col].str.replace('\s+', ' ', regex=True)  # multiple spaces -> single space
data.head()
```

```python
for col in text_cols:
    data[col] = data[col].str.title()
data
```

```python
import re

for col in text_cols:
    data[col] = data[col].str.replace('[^a-zA-Z0-9\s]', '', regex=True)
data
```

```python
print(data['City'].unique()[:20])     # pehle 20 unique cities
print(data['State'].unique()[:20])
print(data['Country'].unique())
```

```python
#ðŸ”¸ Outliers & Invalid Values
# Check if any negative or zero sales exist
data[data['Sales'] <= 0]
```

```python
# Calculate Q1 and Q3
Q1 = data['Sales'].quantile(0.25)
Q3 = data['Sales'].quantile(0.75)
IQR = Q3 - Q1

# Define outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Detect outliers
outliers = data[(data['Sales'] < lower_bound) | (data['Sales'] > upper_bound)]
outliers.shape
```

```python
import matplotlib.pyplot as plt

plt.boxplot(data['Sales'])
plt.title('Sales Outlier Detection (Boxplot)')
plt.ylabel('Sales Amount')
plt.show()
```

```python
# Step 2: Replace outliers with boundary values
import numpy as np
data['Sales_capped'] = np.where(data['Sales'] < lower_bound, lower_bound, data['Sales'])
data['Sales_capped'] = np.where(data['Sales_capped'] > upper_bound, upper_bound, data['Sales_capped'])

# Step 3: Compare before vs after
print("\nBefore Capping:")
print(data['Sales'].describe())

print("\nAfter Capping:")
print(data['Sales_capped'].describe())
```

```python
data.columns
```

```python
#ðŸ”¸ Categorical & Numerical Validation
# Check Postal Code format
print("Unique Postal Code count:", data['Postal Code'].nunique())

# Check for invalid values (0, negative, or very small numbers)
invalid_postal = data[(data['Postal Code'] <= 0) | (data['Postal Code'].isnull())]
print("Invalid Postal Codes:")
print(invalid_postal[['Postal Code']])
```

```python
cat_cols = ['Ship Mode', 'Segment', 'Region']

for col in cat_cols:
    print(f"\nUnique values in {col}:")
    print(data[col].unique())
```

```python
for col in cat_cols:
    data[col] = data[col].str.strip().str.title()
data[col].tail()
```
